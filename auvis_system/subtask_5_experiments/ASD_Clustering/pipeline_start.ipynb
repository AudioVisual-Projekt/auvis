Um einen Durchlauf zu starten einfach das Formular am Ende dieses Notebooks benutzen. (nicht den Code verändern!) 
Daraufhin wird unter "ASD_Clustering/data-bin" ein Ordner "_output" erzeugt, der die entsprechenden Ergebnisse beinhaltet.

```python
import ipywidgets as widgets
from IPython.display import display
from auvis_system.subtask_5_experiments.ASD_Clustering.script.main import main
```

```python
# ==========================================
# 1. Definition der UI-Komponenten (Widgets)
# ==========================================

# Datasets
w_datasets = widgets.SelectMultiple(
    options=['dev', 'train', 'train_lr-asd-2epoch', 'dev_lr-asd-2epoch', 'train_lr-asd-5epoch', 'dev_lr-asd-5epoch', 'eval'],
    value=['dev'],
    description='Datasets:',
    rows=7
)

# --- Segmentation Parameters ---
w_onset = widgets.Text(
    value='3.0', description='Onset:'
)
w_offset = widgets.Text(
    value='3.7', description='Offset:'
)
w_min_duration_on = widgets.Text(
    value='0.7', description='Min Dur On:'
)
w_min_duration_off = widgets.Text(
    value='1.57', description='Min Dur Off:'
)
w_max_chunk_size = widgets.Text(
    value='10', description='Max Chunk:'
)
w_min_chunk_size = widgets.Text(
    value='1', description='Min Chunk:'
)

# --- Clustering Parameters ---
w_threshold = widgets.Text(
    value='0.85', description='Threshold:'
)
w_linkage = widgets.Text(
    value='complete', description='Linkage:'
)

# Button und Output
btn_run = widgets.Button(
    description='Run Experiment',
    button_style='success',
    icon='play'
)
out = widgets.Output()

# ==========================================
# 2. Event-Handler und Logik
# ==========================================

def on_button_clicked(b):
    """
    Callback-Funktion zum Einlesen der UI-Parameter und Starten der Pipeline.
    Unterstützt Grid-Search durch kommagetrennte Eingaben in den Textfeldern.
    """
    with out:
        out.clear_output()
        
        try:
            # 1. Floats parsen
            onset_vals = [float(x.strip()) for x in w_onset.value.split(',')]
            offset_vals = [float(x.strip()) for x in w_offset.value.split(',')]
            min_dur_on_vals = [float(x.strip()) for x in w_min_duration_on.value.split(',')]
            min_dur_off_vals = [float(x.strip()) for x in w_min_duration_off.value.split(',')]
            thresh_vals = [float(x.strip()) for x in w_threshold.value.split(',')]
            
            # 2. Integers parsen
            max_chunk_vals = [int(x.strip()) for x in w_max_chunk_size.value.split(',')]
            min_chunk_vals = [int(x.strip()) for x in w_min_chunk_size.value.split(',')]
            
            # 3. Strings parsen
            linkage_vals = [str(x.strip()) for x in w_linkage.value.split(',')]
            
            print(f"Starting pipeline with datasets: {list(w_datasets.value)}")
            print("-" * 50)
            
            # Parameter-Dictionaries zusammenstellen
            custom_seg = {
                'onset': onset_vals,
                'offset': offset_vals,
                'min_duration_on': min_dur_on_vals,
                'min_duration_off': min_dur_off_vals,
                'max_chunk_size': max_chunk_vals,
                'min_chunk_size': min_chunk_vals
            }
            custom_clust = {
                'threshold': thresh_vals,
                'linkage': linkage_vals
            }
            
            # Pipeline starten
            main(
                datasets=list(w_datasets.value),
                custom_seg_params=custom_seg,
                custom_clust_params=custom_clust
            )
            
            print("-" * 50)
            print("Pipeline execution completed successfully.")
            
        except ValueError as e:
            print(f"Input Error: Please check your formatting (e.g., numbers vs. text). Details: {e}")
        except Exception as e:
            print(f"Pipeline Error: An unexpected error occurred. Details: {e}")

btn_run.on_click(on_button_clicked)


# ==========================================
# 3. Initialisierung der Benutzeroberfläche
# ==========================================

# UI in vertikalen/horizontalen Boxen anordnen für ein sauberes Layout
ui_seg = widgets.VBox([
    widgets.HTML("<b>Segmentation Parameters</b>"), 
    w_onset, w_offset, w_min_duration_on, w_min_duration_off, w_max_chunk_size, w_min_chunk_size
])

ui_clust = widgets.VBox([
    widgets.HTML("<b>Clustering Parameters</b>"), 
    w_threshold, w_linkage
])

ui_layout = widgets.HBox([w_datasets, ui_seg, ui_clust])

# Rendern
display(ui_layout, btn_run, out)
```
