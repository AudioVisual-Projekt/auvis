```python
import os
import glob
import csv

# Base directory where your evaluation_summary_*.csv files live
# Change this if needed
base_dir = "/home/jokai001/auvis/team_b/mcorec_baseline/data-bin/dev"

# Pattern for all summary CSVs
summary_pattern = os.path.join(base_dir, "evaluation_summary_*.csv")
merged_csv_path = os.path.join(base_dir, "evaluation_summary_merged.csv")

# Find all matching files, excluding the merged file itself (if it exists)
summary_files = sorted(
    f for f in glob.glob(summary_pattern)
    if os.path.basename(f) != os.path.basename(merged_csv_path)
)

print(f"Found {len(summary_files)} summary file(s):")
for f in summary_files:
    print(" -", os.path.basename(f))

if not summary_files:
    print("No evaluation_summary_*.csv files found. Nothing to merge.")
else:
    with open(merged_csv_path, "w", newline="") as merged_file:
        writer = csv.writer(merged_file)

        # Write unified header (adds source_csv)
        writer.writerow([
            "source_csv",
            "variant_name",
            "avg_conversation_clustering_f1",
            "avg_speaker_wer",
            "avg_joint_asr_clustering_error",
        ])

        for path in summary_files:
            with open(path, "r", newline="") as f_in:
                reader = csv.reader(f_in)
                header = next(reader, None)  # skip per-file header
                for row in reader:
                    if not row:
                        continue
                    # Prepend the source filename
                    writer.writerow([os.path.basename(path)] + row)

    print(f"\nMerged CSV written to: {merged_csv_path}")
```
