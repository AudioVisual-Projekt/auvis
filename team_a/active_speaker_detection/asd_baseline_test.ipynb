```python
from src.talking_detector import segmentation
import json

asd_path = f"{output_path}/{track_name}_asd.json"

with open(asd_path, "r") as f:
    asd = json.load(f)
    
segments = segmentation.segment_by_asd(asd)
print(segments[0])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame.from_dict(asd, orient='index', columns=["value"])
df["probability"] = 1 / (1 + np.exp(-df["value"]))
ax = df["value"].plot(figsize=(15,7), xlabel="frame", ylabel="asd score")
ax.axhline(y=1.0, color="red", linestyle="--", linewidth=1)

plt.show()
```

```python
ax2 = df["probability"].plot(figsize=(15,7), xlabel="frame", ylabel="asd probability")
ax2.axhline(y=0.5, color="red", linestyle="--", linewidth=1)

plt.show()
```
